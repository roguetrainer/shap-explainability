# LLM Use Case: Interpreting Sentiment Analysis

**Goal:** Understand which words drive a "Positive" or "Negative" sentiment prediction in a small LLM (DistilBERT).


```python
import shap
import transformers
from transformers import pipeline

shap.initjs()
```

```python
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
p = pipeline("text-classification", model=model_name, return_all_scores=True)
text_inputs = ["I really wanted to hate this movie because the actor is annoying, but the plot was actually fantastic."]
```

```python
explainer = shap.Explainer(p)
shap_values = explainer(text_inputs)
```

```python
shap.plots.text(shap_values)
```
